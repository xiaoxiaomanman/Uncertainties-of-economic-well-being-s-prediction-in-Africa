## Load data

````python
df = pd.read_csv('data/dhs_clusters.csv', float_precision='high', index_col=False)
labels = df['wealthpooled'].to_numpy(dtype=np.float32)
locs = df[['lat', 'lon']].to_numpy(dtype=np.float32)
country_labels = df['country'].map(COUNTRIES.index).to_numpy()
````

## 1、Concatenated MS + NL features 16

````python
def ridge_UMAP_ooc_concat_wrapper(model_names: Iterable[str], savedir: str, best_alphas_path: str, emb_method, method_kwargs) -> None:
    '''
    Args
    - model_names: list of str, correspond to keys in MODEL_DIRS (without the fold suffix)
    - savedir: str, path to directory for saving ridge regression weights and predictions
    '''
    features_dict = {}
    for f in FOLDS:
        concat_features = []  # list of np.array, each shape [N, D_i]
        for model_name in model_names:
            model_dir = MODEL_DIRS[f'{model_name}_{f}']
            npz_path = os.path.join(OUTPUTS_ROOT_DIR, 'dhs_ooc', model_dir, 'features.npz')
            npz = load_npz(npz_path, check={'labels': labels})
            concat_features.append(npz['features'])
        concat_features = np.concatenate(concat_features, axis=1)  # shape [N, D_1 + ... + D_m]
        for country in dataset_constants.SURVEY_NAMES[f'DHS_OOC_{f}']['test']:
            features_dict[country] = concat_features

    aa=datetime.datetime.now()#开始时间点
    print('开始时间：',aa)
    ridge_manifold_cv(
        features=features_dict,
        labels=labels,
        group_labels=country_labels,
        group_names=COUNTRIES,
        emb_method = emb_method,
        method_kwargs = method_kwargs,
        do_plot=True,
        savedir=savedir,
        best_alphas_path = best_alphas_path,
        save_weights=True,
        save_dict=dict(locs=locs))
    
    bb=datetime.datetime.now()#结束时间点
    print('结束时间：',bb)
    cc=bb-aa#运行时间，单位是  时:分:秒
    print('运行时间是：',cc)
````

````python
# 16维
model_names = ['resnet_ms', 'resnet_nl']
emb_method = UMAP
method_kwargs = {'n_components': 16, 'random_state': 123}
savedir = os.path.join(OUTPUTS_ROOT_DIR, 'dhs_ooc', 'UMAP', 'resnet_msnl_UMAP','ridge_logo_UMAP')
os.makedirs(savedir,exist_ok=True)
features_dir = os.path.join(OUTPUTS_ROOT_DIR, 'dhs_ooc','UMAP','features_msnl_UMAP/')
os.makedirs(features_dir,exist_ok=True)
best_alphas_path = os.path.join(savedir, 'best_alphas.csv')
ridge_UMAP_ooc_concat_wrapper(model_names, savedir, best_alphas_path, emb_method, method_kwargs)
````

## 2、MS features 16

````python
def ridge_UMAP_ooc_wrapper(model_name: str, savedir: str, best_alphas_path: str, emb_method, method_kwargs) -> None:
    '''
    Args
    - model_name: str, corresponds to keys in MODEL_DIRS (without the fold suffix)
    - savedir: str, path to directory for saving ridge regression weights and predictions
    '''
    features_dict = {}
    for f in FOLDS:
        model_fold_name = f'{model_name}_{f}'
        model_dir = MODEL_DIRS[model_fold_name]
        npz_path = os.path.join(OUTPUTS_ROOT_DIR, 'dhs_ooc', model_dir, 'features.npz')
        npz = load_npz(npz_path, check={'labels': labels})
        features = npz['features']
        for country in dataset_constants.SURVEY_NAMES[f'DHS_OOC_{f}']['test']:
            features_dict[country] = features
    
    aa=datetime.datetime.now()#开始时间点
    print('开始时间：',aa)
    ridge_manifold_cv(
        features=features_dict,
        labels=labels,
        group_labels=country_labels,
        group_names=COUNTRIES,
        emb_method = emb_method,
        method_kwargs = method_kwargs,
        do_plot=True,
        savedir=savedir,
        best_alphas_path = best_alphas_path,
        save_weights=True,
        save_dict=dict(locs=locs))
    
    bb=datetime.datetime.now()#结束时间点
    print('结束时间：',bb)
    cc=bb-aa#运行时间，单位是  时:分:秒
    print('运行时间是：',cc)
````

````python
# 16维
model_name = 'resnet_ms'
emb_method = UMAP
method_kwargs = {'n_components': 16, 'random_state': 123}
savedir = os.path.join(OUTPUTS_ROOT_DIR, 'dhs_ooc', 'UMAP', 'resnet_ms_UMAP','ridge_logo_UMAP')
os.makedirs(savedir,exist_ok=True)
features_dir = os.path.join(OUTPUTS_ROOT_DIR, 'dhs_ooc','UMAP','features_ms_UMAP/')
os.makedirs(features_dir,exist_ok=True)
best_alphas_path = os.path.join(savedir, 'best_alphas.csv')
ridge_UMAP_ooc_wrapper(model_name, savedir, best_alphas_path, emb_method, method_kwargs)
````

````python
# 16维
model_name = 'resnet_nl'
emb_method = UMAP
method_kwargs = {'n_components': 16, 'random_state': 123}
savedir = os.path.join(OUTPUTS_ROOT_DIR, 'dhs_ooc', 'UMAP', 'resnet_nl_UMAP','ridge_logo_UMAP')
os.makedirs(savedir,exist_ok=True)
features_dir = os.path.join(OUTPUTS_ROOT_DIR, 'dhs_ooc','UMAP','features_nl_UMAP/')
os.makedirs(features_dir,exist_ok=True)
best_alphas_path = os.path.join(savedir, 'best_alphas.csv')
ridge_UMAP_ooc_wrapper(model_name, savedir, best_alphas_path, emb_method, method_kwargs)
````
